Считайте датасет из файла train.csv (это данные о выживаемости на Титанике)

```python
train_data = pd.read_csv('/content/train.csv')
```

Посчитайте процент выживаемости у каждого класса пассажиров (Pclass)

```python
for i in range(3):
  print(f"Выживаемость пассажиров {i+1} класса - {round(train_data.groupby('Pclass')['Survived'].sum()[i+1] / len(train_data['Survived']) * 100, 2)}%")
```

```python
train_data['Name'] = train_data['Name'].apply(lambda x: x[x.find('.')+2:])
```

```python
train_data['Name'] = train_data['Name'].apply(lambda x : x if x.find(' ') == -1 else x[:x.find(' ')])
```

```python
train_data['Name']
```

Выведите самое популярное мужское и самое популярное женское имя на корабле

```python
print(f"Самое частое мужское имя - {train_data[train_data['Sex'] == 'male']['Name'].value_counts().index[0]}, оно встречается {train_data[train_data['Sex'] == 'male']['Name'].value_counts()[0]} раз")
print(f"Самое частое женское имя - {train_data[train_data['Sex'] == 'female']['Name'].value_counts().index[0]}, оно встречается {train_data[train_data['Sex'] == 'female']['Name'].value_counts()[0]} раз")
```

Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе

```python
for i in range(3):
  print(f"Самое частое мужское имя {i+1} класса - {train_data[(train_data['Sex'] == 'male') & (train_data['Pclass'] == i+1)]['Name'].value_counts().index[0]}")
  print(f"Самое частое женское имя {i+1} класса - {train_data[(train_data['Sex'] == 'female') & (train_data['Pclass'] == i+1)]['Name'].value_counts().index[0]}")
```

Выведите часть таблицы с пассажирами, возраст которых больше 44 лет

```python
train_data[train_data['Age'] > 44]
```

Выведите часть таблицы с пассажирами, возраст которых меньше 44 лет и которые мужского пола

```python
train_data[(train_data['Age'] < 44) & (train_data['Sex'] == 'male')]
```

Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека)

```python
a = train_data.groupby('Cabin')['PassengerId'].count().values
unique, counts = np.unique(a, return_counts=True)
```

```python
for i, j in zip(unique, counts):
  print(f"{i}-местных кабин {j} шт.")
```
